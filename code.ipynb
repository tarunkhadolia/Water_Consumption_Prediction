```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
# make a column of month 

train_df['month'] = pd.to_datetime(train_df['Timestamp'], dayfirst=True).dt.month
test_df['month'] = pd.to_datetime(test_df['Timestamp'], dayfirst=True).dt.month

# make a column of year
train_df['year'] = pd.to_datetime(train_df['Timestamp'], dayfirst=True).dt.year
test_df['year'] = pd.to_datetime(test_df['Timestamp'], dayfirst=True).dt.year

# make a column of day

train_df['day'] = pd.to_datetime(train_df['Timestamp'], dayfirst=True).dt.day
test_df['day'] = pd.to_datetime(test_df['Timestamp'], dayfirst=True).dt.day

# make a column of hour

train_df['hour'] = pd.to_datetime(train_df['Timestamp'], dayfirst=True).dt.hour
test_df['hour'] = pd.to_datetime(test_df['Timestamp'], dayfirst=True).dt.hour
```

```python
train_df = train_df.drop("Timestamp", axis=1)
test_df = test_df.drop("Timestamp", axis=1)
```

```python
train_df["Temperature"] = train_df["Temperature"].fillna(train_df["Temperature"].mean())
test_df["Temperature"] = test_df["Temperature"].fillna(train_df["Temperature"].mean())
```

```python
train_df.info()
```

```python
train_df['Apartment_Type'] = train_df['Apartment_Type'].fillna(train_df['Apartment_Type'].mode()[0])
train_df['Income_Level'] = train_df['Income_Level'].fillna(train_df['Income_Level'].mode()[0])
train_df['Amenities'] = train_df['Amenities'].fillna(train_df['Amenities'].mode()[0])


test_df['Apartment_Type'] = test_df['Apartment_Type'].fillna(test_df['Apartment_Type'].mode()[0])
test_df['Income_Level'] = test_df['Income_Level'].fillna(test_df['Income_Level'].mode()[0])
test_df['Amenities'] = test_df['Amenities'].fillna(test_df['Amenities'].mode()[0])
```

```python
train_df.info()
test_df.info()
```

```python
train_df['Residents'] = train_df['Residents'].apply(lambda x: 0 if x < 0 else x)
test_df['Residents'] = test_df['Residents'].apply(lambda x: 0 if x < 0 else x)
```

```python
# Calculate mean of water_price

mean_water_price = train_df['Water_Price'].mean()
```

```python
train_df['Guests'] = train_df['Guests'].apply(lambda x: 0 if x < 0 else x)
test_df['Guests'] = test_df['Guests'].apply(lambda x: 0 if x < 0 else x)

train_df['Water_Price'] = train_df['Water_Price'].apply(lambda x: mean_water_price if x < 0 else x)
test_df['Water_Price'] = test_df['Water_Price'].apply(lambda x: mean_water_price if x < 0 else x)
```

```python
# Calculate mean of Peroid_Consumption_Index

mean_period_consumption_index = train_df['Period_Consumption_Index'].mean()

train_df['Period_Consumption_Index'] = train_df['Period_Consumption_Index'].apply(lambda x: mean_period_consumption_index if x < 0 else x)
test_df['Period_Consumption_Index'] = test_df['Period_Consumption_Index'].apply(lambda x: mean_period_consumption_index if x < 0 else x)
```

```python
train_df['Appliance_Usage'] = train_df['Appliance_Usage'].fillna(0)
test_df['Appliance_Usage'] = test_df['Appliance_Usage'].fillna(0)

```

```python
train_df.info()
test_df.info()
```

```python
# Replace valid income levels with corresponding integers
train_df['Income_Level'] = train_df['Income_Level'].replace({'Rich': 1, 'Upper Middle': 2, 'Middle': 3, 'Low': 4})
test_df['Income_Level'] = test_df['Income_Level'].replace({'Rich': 1, 'Upper Middle': 2, 'Middle': 3, 'Low': 4})

train_df['Income_Level'] = pd.to_numeric(train_df['Income_Level'], errors='coerce').fillna(train_df['Income_Level'].mode()[0]).astype(int)
test_df['Income_Level'] = pd.to_numeric(test_df['Income_Level'], errors='coerce').fillna(test_df['Income_Level'].mode()[0]).astype(int)

# Display value counts
train_df['Income_Level'].value_counts()
test_df['Income_Level'].value_counts()
```

```python
# Apply one hot encoding to the categorical columns

df_pandas = pd.get_dummies(train_df, columns=['Apartment_Type', 'Amenities'] , drop_first=True)

# Apply one hot encoding to the categorical columns in test_df

df_test_pandas = pd.get_dummies(test_df, columns=['Apartment_Type', 'Amenities'] , drop_first=True)
```

```python
df_pandas.head()
```

```python
df_test_pandas.head()
```

```python
# Make humidity column as float after cleaning invalid values

# Remove invalid values that cannot be converted to float
df_pandas['Humidity'] = pd.to_numeric(df_pandas['Humidity'].str.rstrip('%'), errors='coerce')

# Fill NaN values (resulting from invalid entries) with the mean of the column
df_pandas['Humidity'] = df_pandas['Humidity'].fillna(df_pandas['Humidity'].mean())

# Remove invalid values that cannot be converted to float

df_test_pandas['Humidity'] = pd.to_numeric(df_test_pandas['Humidity'].str.rstrip('%'), errors='coerce')

# Fill NaN values (resulting from invalid entries) with the mean of the column
df_test_pandas['Humidity'] = df_test_pandas['Humidity'].fillna(df_test_pandas['Humidity'].mean())
```

```python
df_pandas.info()
df_test_pandas.info()
```

```python
# Add a new column 'Total_Occupants' to the training DataFrame
df_pandas['Total_Occupants'] = df_pandas['Residents'] + df_pandas['Guests']
# Add a new column 'Total_Occupants' to the test DataFrame
df_test_pandas['Total_Occupants'] = df_test_pandas['Residents'] + df_test_pandas['Guests']

```

```python
# make test and train

X_train = df_pandas.drop(columns=['Water_Consumption'])
y_train = df_pandas['Water_Consumption']
X_test = df_test_pandas

```

```python
X_train.dtypes
```

```python
# Apply XGBoost Regressor
from xgboost import XGBRegressor

xgb_model = XGBRegressor(max_depth= 4, learning_rate=0.01 , n_estimators=3500, random_state=42)
xgb_model.fit(X_train, y_train)

# make predictions on the test set
y_pred_xgb = xgb_model.predict(X_test)

# save the predictions to a CSV file and  index timestamp
submission_df_xgb = pd.DataFrame({'Water_Consumption': y_pred_xgb})

submission_df_xgb.to_csv('submission_xgb1.csv', index=True)
# Apply LGBM Regressor

```
